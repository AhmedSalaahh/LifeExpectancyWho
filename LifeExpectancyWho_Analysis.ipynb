```python
for col in df.columns:
    if df[col].isnull().sum() > 0:
        df[col] = df[col].fillna(df[col].mean())
```

```python
deaths = df[['Adult Mortality','infant deaths','under-five deaths ']]
immunization = df[['Hepatitis B','Polio','Diphtheria ']]
diseases = df[['Measles ',' HIV/AIDS']]
```

```python
matrix = pd.concat([immunization,diseases],axis=1)
matrix_ex= pd.concat([matrix,df['Life expectancy ']],axis=1)
correlation_matrix = matrix_ex.corr()['Life expectancy ']

print(correlation_matrix)
```

```python
dff=df.drop(['Country','Year', 'Status'],axis=1)
correlations = dff.corr()['Adult Mortality'].sort_values(ascending=False)
print(correlations)
```

```python
sns.kdeplot(data=df[df["Status"] == "Developing"], x="Life expectancy ", fill=True)
sns.kdeplot(data=df[df["Status"] == "Developed"], x="Life expectancy ", fill=True)
plt.title("Life Expectancy Distribution")
legend = ["Developing", "Developed"]
plt.legend(legend, loc='upper left')
plt.show()
```

```python
X = dff.drop(['Life expectancy '],axis=1)
y = dff['Life expectancy ']
```

```python
from sklearn.model_selection import train_test_split
# split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=.8, random_state=42)
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

rf= RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2= r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

```python
# Create a pd.Series of features importances
importances = pd.Series(data=rf.feature_importances_,
                        index= X.columns)

# Sort importances
importances_sorted = importances.sort_values(ascending=False).head(10)

# Draw a horizontal barplot of importances_sorted
importances_sorted.plot(kind='barh')
plt.title('Features Importances')
plt.show()
```
